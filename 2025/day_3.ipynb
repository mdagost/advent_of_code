```python
TEST_INFILE = "inputs/day_3_test_1.txt"
INFILE = "inputs/day_3.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
def get_max_pair(num):
    max_pair = float("-inf")
    for i in range(len(num)-1):
        for j in range(i+1, len(num)):
            pair = int(f"{num[i]}{num[j]}")
            if pair > max_pair:
                max_pair = pair
    return max_pair
```

```python
sum(get_max_pair(l) for l in lines)
```

## Part 2

```python
def get_msb(num, digits_left):
    if digits_left == 1:
        return max(d for d in num),len(num)

    best_value = float("-inf")
    for msb in range(9, 0, -1):
        pos = num.find(str(msb))
        # we found something, and we leave enough room for the rest of the number
        if pos > -1 and (len(num) - 1 - pos) >= (digits_left - 1):
            value = msb * 10**digits_left
            if value > best_value:
                best_value = value
                best_msb = msb
                best_pos = pos
    return best_msb, best_pos
```

```python
def get_max_combo(num):
    combo = []
    digits_left = 12
    while digits_left > 0:
        digit, pos = get_msb(num, digits_left)
        combo.append(str(digit))
        num = num[pos+1:]
        digits_left -= 1

    combo = "".join(combo)
    assert len(combo) == 12

    return int(combo)
```

```python
sum(get_max_combo(l) for l in lines)
```
