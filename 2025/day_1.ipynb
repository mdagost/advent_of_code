## Part 1

```python
import operator
```

```python
TEST_INFILE = "inputs/day_1_test_1.txt"
INFILE = "inputs/day_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
ops = {
    "L": operator.sub,
    "R": operator.add,
}
```

```python
lines = [(ops[l[0]], int(l[1:])) for l in lines]
```

```python
value = 50
zeros = 0
for line in lines:
    op, num = line
    value = op(value, num) % 100
    if value == 0:
        zeros += 1
```

## Part 2

```python
value = 50
zeros = 0
for line in lines:
    op, num = line

    #print(f"Starting at {value} going {'L' if op == operator.sub else 'R'}")

    full_circles = num // 100
    
    if full_circles > 0:
        zeros += full_circles
        num = num - full_circles * 100

    start_value = value
    full_value = op(value, num)
    value = op(value, num) % 100

    #print(f"Ending at {value}")

    if value == 0:
        zeros += 1
    elif start_value > 0 and op == operator.add and full_value > 99:
        #print("Crossed zero!")
        zeros += 1
    elif start_value > 0 and op == operator.sub and full_value < 0:
        #print("Crossed zero!")
        zeros += 1

    #print(value)
```
