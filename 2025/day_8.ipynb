```python
from itertools import combinations
```

```python
TEST_INFILE = "inputs/day_8_test_1.txt"
INFILE = "inputs/day_8.txt"

with open(TEST_INFILE) as infile:
#with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
boxes = [tuple(int(e) for e in line.split(",")) for line in lines]
```

```python
def dist(p1, p2):
    return sum(map(lambda t: (t[0]-t[1])**2, zip(p1, p2)))
```

```python
dists = [(p1, p2, dist(p1, p2)) for p1, p2 in combinations(boxes, 2)]
```

```python
N_SHORTEST = 10
sorted_dists = sorted(dists, key=lambda t: t[2])
```

```python
circuits = []
connected = 0
for p1, p2, _ in sorted_dists:
    print(connected)
    if connected == N_SHORTEST:
        break

    print(f"Looking at {p1}, {p2}")
    
    skipped, added = False, False
    for circuit in circuits:
        if p1 in circuit and p2 in circuit:
            skipped = True
            print("Already connected!")
            break

        if p1 in circuit or p2 in circuit:
            print(f"Adding it to circuit {circuit}")
            circuit |= set([p1, p2])
            added = True
            connected += 1
            print(circuit)
            break
    
    if not skipped and not added:
        circuits.append(set([p1, p2]))
        connected += 1
        print(f"Adding it to its own circuit.")
        print(circuits)
    print(circuits)
    print("\n")
```

```python
for c in circuits:
    print(c)
```
