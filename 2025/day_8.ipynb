## Part 1

```python
import operator
from functools import reduce
from itertools import combinations
```

```python
TEST_INFILE = "inputs/day_8_test_1.txt"
INFILE = "inputs/day_8.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()

#N_SHORTEST = 10
N_SHORTEST = 1000
```

```python
boxes = [tuple(int(e) for e in line.split(",")) for line in lines]
```

```python
def dist(p1, p2):
    return sum(map(lambda t: (t[0]-t[1])**2, zip(p1, p2)))
```

```python
dists = [(p1, p2, dist(p1, p2)) for p1, p2 in combinations(boxes, 2)]
```

```python
sorted_dists = sorted(dists, key=lambda t: t[2])[:N_SHORTEST]
```

```python
circuits = []
for p1, p2, _ in sorted_dists:
    this_pair = set([p1, p2])
    
    to_merge = list(filter(lambda c: p1 in c or p2 in c, circuits))

    if len(to_merge) == 0:
        circuits.append(this_pair)
    else:
        new_circuit = set()
        for c in to_merge:
            m = circuits.pop(circuits.index(c))
            new_circuit |= m
        new_circuit |= set([p1, p2])
        circuits.append(new_circuit)
```

```python
reduce(operator.mul, sorted(len(c) for c in circuits)[::-1][0:3])
```
