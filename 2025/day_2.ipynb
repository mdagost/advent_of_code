## Part 1

```python
TEST_INFILE = "inputs/day_2_test_1.txt"
INFILE = "inputs/day_2.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read()

ranges = [l.split("-") for l in lines.split(",")]
ranges = [[int(e) for e in r] for r in ranges]
```

```python
def is_invalid(num):
    num = str(num)
    
    if len(num) % 2 != 0:
        return False
    
    mid = len(num) // 2
    if num[:mid] == num[mid:]:
        return True
    
    return False
```

```python
invalid = []
for rl, ru in ranges:
    for i in range(rl, ru+1):
        if is_invalid(i):
            invalid.append(i)
```

```python
sum(invalid)
```

## Part 2

```python
def check_motif(num, size):
    num = str(num)
    
    if len(num) % size != 0:
        return False
    
    chunks = set()
    lower, upper = 0, size
    n_motifs = len(num) // size
    for i in range(n_motifs):
        chunks.add(num[lower:upper])
        lower += size
        upper += size
        if len(chunks) > 1:
            return False
        
    return True


def is_invalid(num):
    num = str(num)

    max_motif_size = len(num) // 2

    for motif_size in range(1, max_motif_size + 1):
        if check_motif(num, motif_size):
            return True
    return False
```

```python
invalid = []
for rl, ru in ranges:
    for i in range(rl, ru+1):
        if is_invalid(i):
            invalid.append(i)
```

```python
sum(invalid)
```
