```python
TEST_INFILE = "inputs/day_5_test_1.txt"
INFILE = "inputs/day_5.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    rs, ids = infile.read().split("\n\n")
    rs = rs.splitlines()
    ids = ids.splitlines()
    ids = [int(i) for i in ids]
```

```python
class Range:
    def __init__(self, rmin, rmax):
        assert rmin <= rmax
        self.rmin = rmin
        self.rmax = rmax


    def contains(self, num):
        if self.rmin <= num and num <= self.rmax:
            return True
        return False
```

```python
ranges = []
for r in rs:
    rmin, rmax = r.split("-")
    ranges.append(Range(int(rmin), int(rmax)))
```

```python
def is_fresh(_id):
    for r in ranges:
        if r.contains(_id):
            return True
    return False
```

```python
len(list(filter(lambda i: is_fresh(i), ids)))
```

```python
from functools import reduce
```

```python
class Range:
    def __init__(self, rmin, rmax):
        assert rmin <= rmax
        self.rmin = rmin
        self.rmax = rmax


    def __repr__(self):
        return f"[{self.rmin}, {self.rmax}]"


    def __len__(self):
        return self.rmax - self.rmin + 1    


    def contains(self, num):
        if self.rmin <= num and num <= self.rmax:
            return True
        return False


    def intersects(self, other):
        if self.contains(other.rmin) or self.contains(other.rmax):
            return True
        return False
    

    def combine(self, other):
        if not self.intersects(other):
            return None
        rmin = min(self.rmin, other.rmin) 
        rmax = max(self.rmax, other.rmax)
        return Range(rmin, rmax)
```

```python
ranges = []
for r in rs:
    rmin, rmax = r.split("-")
    ranges.append(Range(int(rmin), int(rmax)))
```

```python
ranges = sorted(ranges, key=lambda r: r.rmin)
```

```python
idx = 0
last = ranges[-1]
last_checked = None
while not last_checked or last_checked != last:
    r1 = ranges[idx]
    r2 = ranges[idx+1]
    if not r1.combine(r2):
        #print(f"{r1} and {r2} don't interset.")
        idx += 1
    else:
        #print(f"{r1} and {r2} DO interset.")
        _ = ranges.pop(idx+1)
        _ = ranges.pop(idx)
        ranges = ranges[:idx] + [r1.combine(r2)] + ranges[idx:]
    last_checked = r2
```

```python
sum(len(r) for r in ranges)
```
