```python
TEST_INFILE = "inputs/day_12_test_1.txt"
INFILE = "inputs/day_12.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().split("\n\n")
    
    presents = [l.split()[1:] for l in lines[:-1]]
    
    trees = lines[-1].split("\n")
    for n, tree in enumerate(trees.copy()):
        grid = tree.split(": ")[0]
        tree_presents = [int(e) for e in tree.split(": ")[1].split()]
        trees[n] = (grid, tree_presents)
```

```python
def count_occupied(p):
    return sum(1 if c == "#" else 0 for r in p for c in r)

def grid_size(grid):
    return int(grid.split("x")[0]) * int(grid.split("x")[1])

def draw_present(p):
    for row in p:
        print(row)

def vflip(p):
    return p[::-1]

def hflip(p):
    return [r[::-1] for r in p]

def rotate(p):
    # transpose and then reverse
    out = [[0]*3, [0]*3, [0]*3]
    for r_n, r in enumerate(p):
        for c_n, c in enumerate(r):
            out[c_n][r_n] = c
    return ["".join(r)[::-1] for r in out]
```

```python
counts = [count_occupied(p) for p in presents]
```

```python
valid_trees = []
for tree in trees:
    grid, ps = tree
    occupied = 0
    for a, b in zip(ps, counts):
        occupied += a*b
    if occupied > grid_size(grid):
        continue
    else:
        print(f"{grid}: {grid_size(grid)} available, {occupied} needed.")
        valid_trees.append(tree)
```
