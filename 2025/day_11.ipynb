## Part 1

```python
from functools import lru_cache
```

```python
TEST_INFILE = "inputs/day_11_test_1.txt"
TEST_INFILE_2 = "inputs/day_11_test_2.txt"
INFILE = "inputs/day_11.txt"

#with open(TEST_INFILE) as infile:
#with open(TEST_INFILE_2) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
edges = {}
for line in lines:
    node = line.split(":")[0]
    assert node not in edges

    neighbors = line.split(":")[1].strip().split()
    edges[node] = neighbors
```

```python
@lru_cache(maxsize=None)
def count_paths(node, goal="out"):
    paths = 0
    if node not in edges:
        return 0

    for e in edges[node]:
        if e != goal:
            paths += count_paths(e, goal)
        else:
            paths += 1
    return paths
```

```python
count_paths("you")
```
