## Part 1

```python
import operator
from functools import reduce
```

```python
TEST_INFILE = "inputs/day_6_test_1.txt"
INFILE = "inputs/day_6.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
operations = {
    "+": operator.add,
    "*": operator.mul,
}
```

```python
nums = [[int(e) for e in line.split()] for line in lines[:-1]]
ops = lines[-1].split()
```

```python
answers = []
for problem_num, problem in enumerate(zip(*nums)):
    answers.append(reduce(operations[ops[problem_num]], problem))
```

```python
sum(answers)
```

## Part 2

```python
TEST_INFILE = "inputs/day_6_test_1.txt"
INFILE = "inputs/day_6.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
operations = {
    "+": operator.add,
    "*": operator.mul,
}
```

```python
problems = []
problem = []
for col in zip(*lines):
    # all spaces is our problem separator
    if sum(len(e.strip()) for e in col) == 0:
        problems.append(problem)
        problem = []
    else:
        problem.append(col)
problems.append(problem)
```

```python
answers = []
for problem in problems:
    op = problem[0][-1]
    problem[0] = problem[0][:-1]
    problem = [int("".join(n).strip()) for n in problem]
    answers.append(reduce(operations[op], problem))
```

```python
sum(answers)
```
