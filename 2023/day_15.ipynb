```python
def HASH(s):
    value = 0
    for char in s:
        value += ord(char)
        value *= 17
        value = value % 256
    return value
```

```python
import re
```

```python
def process_step(step):
    parts  = re.split(r"(=|-)", step)
    label  = parts[0]
    box    = HASH(label)
    op     = parts[1]
    focal  = parts[2]
    lenses = hashmap[box]
    
    print(f"Step {step} deals with Box {box} which has lenses {lenses}")

    matching_lens = [n for n, l in enumerate(lenses) if l[0] == label]
    print(f"\t matching lens is {matching_lens}")
    if op == "-":
        if len(matching_lens) == 1:
            # remove it and shift everything else left
            new_lenses = lenses[:matching_lens[0]] + lenses[matching_lens[0]+1:]
            hashmap[box] = new_lenses
        elif len(matching_lens) > 1:
            raise ValueError("Multiple matching lenses found!!")
    elif op == "=":
        if len(matching_lens) == 1:
            # replace it with our new focal length
            new_lenses = lenses[:matching_lens[0]] + [(label, focal)] + lenses[matching_lens[0]+1:]
            hashmap[box] = new_lenses
        elif len(matching_lens) == 0:
            # add it to the end
            lenses.append((label, focal))
        else:
            raise ValueError("Multiple matching lenses found!!")

    print(f"\tAfter step {step} Box {box} has lenses {hashmap[box]}")
```

```python
hashmap = {n: [] for n in range(256)}
for step in steps:
    process_step(step)
```

```python
def get_focusing_power(hashmap):
    focusing_power = 0
    for box, lenses in hashmap.items():
        for n, l in enumerate(lenses):
            focusing_power += (box + 1) * (n + 1) * int(l[1])
    return focusing_power
```
