```python
class Point:
    def __init__(self, row, col):
        self.row = row
        self.col = col

    def __repr__(self):
        return f"({self.row}, {self.col})"

    def __add__(self, other):
        return Point(self.row + other.row, self.col + other.col)

    def __radd__(self, other):
        return self + other
    
    def __eq__(self, other):
        return self.row == other.row and self.col == other.col
    
    def __hash__(self):
        return hash((self.row, self.col))
    

class Direction(Enum):
    UP = Point(-1, 0)
    DOWN = Point(1, 0)
    LEFT = Point(0, -1)
    RIGHT = Point(0, 1)
```

```python
def get_all_neighbors(point, grid):
    neighbors = []
    up    = point + Direction.UP.value
    down  = point + Direction.DOWN.value
    left  = point + Direction.LEFT.value
    right = point + Direction.RIGHT.value

    if up.row >= 0:
        neighbors.append(("N", up))
    if down.row < len(grid):
        neighbors.append(("S", down))
    if left.col >= 0:
        neighbors.append(("W", left))
    if right.col < len(grid[0]):
        neighbors.append(("E", right))

    return neighbors

def get_all_neighbors_symbols(point, grid):
    neighbors = get_all_neighbors(point, grid)
    return [(direction, grid[n.row][n.col], n) for direction, n in neighbors]
```

```python
assert sorted(get_all_neighbors(Point(0, 0), grid)) == [("E", Point(0, 1)), ("S", Point(1, 0))]
assert sorted(get_all_neighbors(Point(1, 0), grid)) == [("E", Point(1, 1)), ("N", Point(0, 0)), ("S", Point(2, 0))]
assert sorted(get_all_neighbors(Point(1, 1), grid)) == [("E", Point(1, 2)), ("N", Point(0, 1)), ("S", Point(2, 1)), ("W", Point(1, 0))]
assert sorted(get_all_neighbors(Point(len(grid), len(grid[0])), grid)) == [("N", Point(len(grid) - 1, len(grid[0]))), ("W", Point(len(grid), len(grid[0]) - 1))]
```

```python
connections = {
    "N": {
        "|": (Direction.DOWN.value, Direction.UP.value, "N"),
        "7": (Direction.DOWN.value, Direction.LEFT.value, "W"),
        "F": (Direction.DOWN.value, Direction.RIGHT.value, "E")
    },
    "S": {
        "|": (Direction.UP.value, Direction.DOWN.value, "S"),
        "L": (Direction.UP.value, Direction.RIGHT.value, "E"),
        "J": (Direction.UP.value, Direction.LEFT.value, "W")
    },
    "E": {
        "-": (Direction.LEFT.value, Direction.RIGHT.value, "E"),
        "7": (Direction.LEFT.value, Direction.DOWN.value, "S"),
        "J": (Direction.LEFT.value, Direction.UP.value, "N")
    },
    "W": {
        "-": (Direction.RIGHT.value, Direction.LEFT.value, "W"),
        "L": (Direction.RIGHT.value, Direction.UP.value, "N"),
        "F": (Direction.RIGHT.value, Direction.DOWN.value, "S")
    }
}
```

```python
def get_starting_neighbor(start, grid):
    neighbors = []
    for direction, symbol, pos in get_all_neighbors_symbols(start, grid):
        #print(f"Checking direction {direction} with symbol {symbol}")
        if symbol in connections[direction]:
            #print("Connnects.")
            neighbors.append((direction, pos))
    assert len(neighbors) == 2
    # just pick the first
    return neighbors[0]
```

```python
for line in lines:
    print("".join(line))
```

```python
edges = {}
direction, pos = get_starting_neighbor(start, grid)

steps = 1
while pos != start:
    symbol = grid[pos.row][pos.col]
    print(f"Current position: {pos}, symbol: {symbol}.  Came here via a {direction} move.")

    # make our edges
    prev_pos, next_pos, next_direction = connections[direction][symbol]
    edges[prev_pos + pos] = pos
    edges[pos] = next_pos + pos

    steps += 1
    pos = next_pos + pos
    direction = next_direction
```

```python
visited = set()
outside = set()

loop_nodes = set(edges.keys()).union(edges.values())
top_edge    = set(Point(row, col) for row, col in zip(repeat(0), range(len(grid[0]))) if Point(row, col) not in loop_nodes)
bottom_edge = set(Point(row, col) for row, col in zip(repeat(len(grid) - 1), range(len(grid[0]))) if Point(row, col) not in loop_nodes)
left_edge   = set(Point(row, col) for row, col in zip(range(len(grid)), repeat(0)) if Point(row, col) not in loop_nodes)
right_edge  = set(Point(row, col) for row, col in zip(range(len(grid)), repeat(len(grid[0]) - 1)) if Point(row, col) not in loop_nodes)

outside.update(top_edge)
outside.update(bottom_edge)
outside.update(left_edge)
outside.update(right_edge)

while len(outside.difference(visited)) > 0:
    current = list(outside.difference(visited))[0]
    visited.add(current)
    
    for direction, neighbor in get_all_neighbors(current, grid):
        if neighbor not in visited and neighbor not in loop_nodes:
            outside.add(neighbor)
```

```python
inside = 0
html = ""
for row, line in enumerate(lines):
    for col, c in enumerate(line):
        if Point(row, col) in outside:
            html += '<span style="background-color:gray;font-weight:bold;font-family:monospace;">O</span>'
        elif Point(row, col) in loop_nodes:
            html += f'<span style="font-family:monospace;">{c}</span>'
        else:
            inside += 1
            html += '<span style="background-color:red;font-weight:bold;color:black;font-family:monospace;">I</span>'
    html += "</br>"
```

```python
all_points = set(Point(row, col) for row in range(len(grid)) for col in range(len(grid[0])))
inside_candidates = all_points.difference(loop_nodes).difference(outside)
```

```python
def get_outside_point(point):
    o_col = 0
    while o_col < len(grid[point.row]) and Point(point.row, o_col) not in outside:
        o_col += 1
    
    if o_col < len(grid[point.row]):
        return "|", Point(point.row, o_col)

    o_row = 0
    while o_row < len(grid) and Point(o_row, point.col) not in outside:
        o_row += 1

    if o_row < len(grid):
        return "-", Point(o_row, point.col)

    raise ValueError(f"Could not find outside point for {point}")
```

```python
def ccw(A, B, C):
    return (C.col - A.col) * (B.row - A.row) > (B.col - A.col) * (C.row - A.row)

def segments_intersect(A, B, C, D):
    return ccw(A, C, D) != ccw(B, C, D) and ccw(A, B, C) != ccw(A, B, D)

def count_crossings(A, B, edges):
    crossings = 0
    for P1, P2 in edges.items():
        if segments_intersect(A, B, P1, P2):
            crossings += 1
    return crossings
```

```python
for c in inside_candidates:
    cross_symbol, outside_point = get_outside_point(c)
    #print(f"Outside point for {c}: {outside_point} with {cross_symbol}")
    if outside_point:
        n_crossings = count_crossings(c, outside_point, edges)
        #print(n_crossings)

        if n_crossings % 2 == 0:
            outside.add(c)
```
