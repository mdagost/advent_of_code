## Part 1

```python
from collections import Counter
from itertools import groupby
```

```python
TEST_INFILE = "inputs/day_7_test_1.txt"
INFILE = "inputs/day_7_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()

hands = [line.split() for line in lines]
hands = [[h[0], int(h[1])] for h in hands]
```

```python
hands
```

```python
cards = ["A", "K", "Q", "J", "T", "9", "8", "7", "6", "5", "4", "3", "2"]
card_strengths = {c: s for c, s in zip(cards, list(range(1, len(cards)+1)))}
```

```python
card_strengths
```

```python
def classify_hand(hand):
    counts = sorted(Counter(hand[0]).values())
    
    if counts == [1, 1, 1, 1, 1]:
        return "g: high card"
    elif counts == [1, 1, 1, 2]:
        return "f: one pair"
    elif counts == [1, 2, 2]:
        return "e: two pair"
    elif counts == [1, 1, 3]:
        return "d: three of a kind"
    elif counts == [2, 3]:
        return "c: full house"
    elif counts == [1, 4]:
        return "b: four of a kind"
    elif counts == [5]:
        return "a: five of a kind"
```

```python
results = sorted([(h[0], h[1], classify_hand(h)) for h in hands], key=lambda x: x[2])
```

```python
results
```

```python
def card_strength_sort(h):
    return [card_strengths[c] for c in h[0]]

ranks = []
for hand_type, hand_list in groupby(results, key=lambda t: t[2]):
    ranks.append((hand_type, sorted(list(hand_list), key=lambda t: card_strength_sort(t))))
ranks
ranks = sorted(ranks, key=lambda t: t[0])
```

```python
ranks
```

```python
flattened = [hand for ranked_hands in ranks for hand in ranked_hands[1]]
reversed = flattened[::-1]
```

```python
total = 0
for rank, hand in enumerate(reversed, start=1):
    #sprint(f"{rank:2}: {hand[0]}:{hand[1]} ({hand[2]})")
    total += hand[1] * rank

total
```

## Part 2

```python
def classify_hand_with_wilds(hand):
    if "J" not in hand[0]:
        return classify_hand(hand)
    
    else:
        counts = Counter(hand[0])
        jokers = counts.pop("J")
        counts = sorted(counts.values())

        if jokers == 5:
            return "a: five of a kind"
        elif jokers == 4:
            return "a: five of a kind"
        elif jokers == 3:
            if counts == [2]:
                return "a: five of a kind"
            else:
                return "b: four of a kind"
        elif jokers == 2:
            if counts == [3]:
                return "a: five of a kind"
            elif counts == [1, 2]:
                return "b: four of a kind"
            else:
                return "d: three of a kind"
        else:
            if counts == [4]:
                return "a: five of a kind"
            elif counts == [1, 3]:
                return "b: four of a kind"
            elif counts == [2, 2]:
                return "c: full house"
            elif counts == [1, 1, 2]:
                return "d: three of a kind"
            else:
                return "f: one pair"           
```

```python
assert classify_hand_with_wilds(["JJJJJ"]) == "a: five of a kind"

assert classify_hand_with_wilds(["JJJJ2"]) == "a: five of a kind"

assert classify_hand_with_wilds(["JJJ22"]) == "a: five of a kind"
assert classify_hand_with_wilds(["JJJ21"]) == "b: four of a kind"

assert classify_hand_with_wilds(["JJ222"]) == "a: five of a kind"
assert classify_hand_with_wilds(["JJ225"]) == "b: four of a kind"
assert classify_hand_with_wilds(["JJ523"]) == "d: three of a kind"


assert classify_hand_with_wilds(["J2222"]) == "a: five of a kind"
assert classify_hand_with_wilds(["J2225"]) == "b: four of a kind"
assert classify_hand_with_wilds(["J2255"]) == "c: full house"
assert classify_hand_with_wilds(["J2245"]) == "d: three of a kind"
assert classify_hand_with_wilds(["J2345"]) == "f: one pair"
```
