## Part 1

```python
TEST_INFILE = "inputs/day_4_test_1.txt"
INFILE = "inputs/day_4_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()

cards = [line.split(":")[1].strip() for line in lines]
cards = [(set(card.split("|")[0].strip().split()), set(card.split("|")[1].strip().split())) for card in cards]
```

```python
points = [2 ** (len(card[0].intersection(card[1])) - 1) for card in cards]
points = list(filter(lambda c: c != 0.5, points))
```

```python
sum(points)
```

## Part 2

```python
from collections import Counter
```

```python
wins = [len(card[0].intersection(card[1])) for card in cards]
```

```python
counts = Counter(range(1, len(cards) + 1))
```

```python
for index, win_count in enumerate(wins):
    #print(counts)
    #print(f"Card {index + 1} has {win_count} wins.")
    cards_to_add = list(range(index + 2, index + 2 + win_count)) * counts[index + 1]
    counts += Counter(cards_to_add)
    #print(cards_to_add)
```

```python
sum(dict(counts).values())
```
