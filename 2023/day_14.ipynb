```python
def sparse_cols_to_rows(map_by_cols, n_rows=len(lines), n_cols=len(lines[0])):
    map_by_rows = []
    for row_num in range(n_rows):
        new_row = []
        for col_num in range(n_cols):
            col = map_by_cols[col_num]
            symbol = [el[1] for el in col if el[0] == row_num]
            if len(symbol) != 0:
                symbol = symbol[0]
                new_row.append((col_num, symbol))
        map_by_rows.append(new_row)
    return map_by_rows


def sparse_rows_to_cols(map_by_rows, n_rows=len(lines), n_cols=len(lines[0])):
    map_by_cols = []
    for col_num in range(n_cols):
        new_col = []
        for row_num in range(n_rows):
            row = map_by_rows[row_num]
            symbol = [el[1] for el in row if el[0] == col_num]
            if len(symbol) != 0:
                symbol = symbol[0]
                new_col.append((row_num, symbol))
        map_by_cols.append(new_col)
    return map_by_cols


def print_sparse_rows(map_by_rows, n_cols=len(lines[0])):
    for row in map_by_rows:
        for col in range(n_cols):
            if col not in [el[0] for el in row]:
                print(".", end="")
            else:
                print([el[1] for el in row if el[0] == col][0], end="")
        print("")
```

```python
map_by_rows = [[(col_n, row[col_n]) for col_n in range(len(lines[0])) if row[col_n] != '.'] for row in lines]
map_by_cols = sparse_rows_to_cols(map_by_rows)
```

```python
assert sparse_cols_to_rows(sparse_rows_to_cols(map_by_rows)) == map_by_rows
```

```python
def tilt_col(col):
    new_col = []
    last_barrier_index = -1
    for el in col:
        if el[1] == "#":
            last_barrier_index = el[0]
            new_col.append(el)
        elif el[1] == "O":
            new_el = (last_barrier_index + 1, "O")
            last_barrier_index = new_el[0]
            new_col.append(new_el)

    return new_col
```

```python
tilted_map_by_cols = [tilt_col(col) for col in map_by_cols]
```

```python
print_sparse_rows(sparse_cols_to_rows(tilted_map_by_cols))
```

```python
def calculate_load(n_rows, map_by_cols):
    total_load = 0
    for col in map_by_cols:
        for el in col:
            if el[1] == "O":
                total_load += (n_rows - el[0])

    return total_load
```

```python
calculate_load(len(lines), tilted_map_by_cols)
```

## Part 2

```python
def tilt_col(col, direction):
    new_col = []
    last_barrier_index = -1 if direction == "N" else len(lines)
    col = col if direction == "N" else col[::-1]
    for el in col:
        if el[1] == "#":
            last_barrier_index = el[0]
            new_col.append(el)
        elif el[1] == "O":
            new_el = (last_barrier_index + 1, "O") if direction == "N" else (last_barrier_index - 1, "O")
            last_barrier_index = new_el[0]
            new_col.append(new_el)

    return new_col if direction == "N" else new_col[::-1]


def tilt_row(row, direction):
    new_row = []
    last_barrier_index = -1 if direction == "W" else len(lines)
    row = row if direction == "W" else row[::-1]
    for el in row:
        if el[1] == "#":
            last_barrier_index = el[0]
            new_row.append(el)
        elif el[1] == "O":
            new_el = (last_barrier_index + 1, "O") if direction == "W" else (last_barrier_index - 1, "O")
            last_barrier_index = new_el[0]
            new_row.append(new_el)

    return new_row if direction == "W" else new_row[::-1]


def cycle(map_by_rows):
    # transpose R => C and tilt N
    map_by_cols = sparse_rows_to_cols(map_by_rows)
    tilted_map_by_cols = [tilt_col(col, "N") for col in map_by_cols]
    
    # transpose C => R and tilt W
    map_by_rows = sparse_cols_to_rows(tilted_map_by_cols)
    tilted_map_by_rows = [tilt_row(row, "W") for row in map_by_rows]
    
    # transpose R => C and tilt S
    map_by_cols = sparse_rows_to_cols(tilted_map_by_rows)
    tilted_map_by_cols = [tilt_col(col, "S") for col in map_by_cols]

    # transpose C => R and tilt E
    map_by_rows = sparse_cols_to_rows(tilted_map_by_cols)
    tilted_map_by_rows = [tilt_row(row, "E") for row in map_by_rows]

    return tilted_map_by_rows
```

```python
results = {tuple(tuple(row) for row in map_by_rows): 0}
tilted = map_by_rows
for N in range(1, 10000):
    tilted = cycle(tilted)
    if tuple(tuple(row) for row in tilted) in results:
       print(f"Found a cycle at {N}!")
       matching_round = results[tuple(tuple(row) for row in tilted)]
       print_sparse_rows(tilted)
       print(f"\n^ should match v from Round {matching_round}\n")
       print_sparse_rows([k for k, v in results.items() if v == matching_round][0])
       break
    results[tuple(tuple(row) for row in tilted)] = N
```

```python
GOAL = 1_000_000_000
remainder = GOAL - (N + (N - matching_round) * ((GOAL - N) // (N - matching_round)))
remainder
```

```python
final_config = [k for k, v in results.items() if v == (matching_round + remainder)]
print_sparse_rows(final_config[0])
```

```python
calculate_load(len(lines), sparse_rows_to_cols(final_config[0]))
```
