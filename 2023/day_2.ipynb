## Part 1

```python
import re
```

```python
TEST_INFILE = "inputs/day_2_test_1.txt"
INFILE = "inputs/day_2_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()

games = [line.split(":") for line in lines]
games = [(int(t[0].replace("Game ", "")), t[1].strip().split("; ")) for t in games]
```

```python
def parse_game(game) -> tuple[int, int, int]:
    m = re.search(r"(\d+) red", game)
    if m:
        red = int(m.group(1))
    else:
        red = 0

    m = re.search(r"(\d+) green", game)
    if m:
        green = int(m.group(1))
    else:
        green = 0

    m = re.search(r"(\d+) blue", game)
    if m:
        blue = int(m.group(1))
    else:
        blue = 0

    return red, green, blue
```

```python
games = [(game[0], [parse_game(g) for g in game[1]]) for game in games]
```

```python
def game_is_possible(game, R, G, B) -> bool:
    for g in game:
        if g[0] > R or g[1] > G or g[2] > B:
            return False
    return True
```

```python
sum([game[0] for game in games if game_is_possible(game[1], 12, 13, 14)])
```

## Part 2

```python
def get_power(game):
    zipped = list(zip(*game[1]))
    return max(zipped[0]) * max(zipped[1]) * max(zipped[2])
```

```python
sum(get_power(g) for g in games)
```
