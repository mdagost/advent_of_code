## Part 1

```python
from itertools import combinations
```

```python
TEST_INFILE = "inputs/day_11_test_1.txt"
INFILE = "inputs/day_11_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
lines
```

```python
empty_rows = [n for n, line in enumerate(lines) if all(c == '.' for c in line)]
empty_cols = [c for c in range(len(lines[0])) if all(lines[r][c] == '.' for r in range(len(lines)))]
galaxies = [(row, col) for row, line in enumerate(lines) for col, c in enumerate(line) if c == "#"]
```

```python
empty_rows, empty_cols
```

```python
len(galaxies)
```

```python
total_dist = 0
for a, b in combinations(galaxies, 2):
    row_start = min(a[0], b[0])
    row_end   = max(a[0], b[0])
    empty_row_crossings = len(set(range(row_start, row_end + 1)).intersection(empty_rows))

    col_start = min(a[1], b[1])
    col_end   = max(a[1], b[1])
    empty_col_crossings = len(set(range(col_start, col_end + 1)).intersection(empty_cols))

    dist = (row_end - row_start) + (col_end - col_start) + empty_row_crossings + empty_col_crossings
    total_dist += dist

total_dist
```

## Part 2

```python
EXPANSION_FACTOR = 1_000_000

total_dist = 0
for a, b in combinations(galaxies, 2):
    row_start = min(a[0], b[0])
    row_end   = max(a[0], b[0])
    empty_row_crossings = len(set(range(row_start, row_end + 1)).intersection(empty_rows))

    col_start = min(a[1], b[1])
    col_end   = max(a[1], b[1])
    empty_col_crossings = len(set(range(col_start, col_end + 1)).intersection(empty_cols))

    dist = (row_end - row_start) + (col_end - col_start) + (EXPANSION_FACTOR - 1)*(empty_row_crossings + empty_col_crossings)
    total_dist += dist

total_dist
```
