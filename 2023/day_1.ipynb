## Part 2

```python
import re
```

```python
TEST_INFILE_2 = "inputs/day_1_test_2.txt"
INFILE = "inputs/day_1_1.txt"

#with open(TEST_INFILE_2) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
words = {
    "one": "1",
    "two": "2",
    "three": "3",
    "four": "4",
    "five": "5",
    "six": "6",
    "seven": "7",
    "eight": "8",
    "nine": "9",
}
```

```python
def convert_to_number_list(line):
    # turn numbers into words
    for word, num in words.items():
        line = line.replace(num, word)

    # get all the word matches
    regex = "|".join(words.keys())
    matches = re.findall(f"(?=({regex}))", line)

    return [words[match] for match in matches]
```

```python
number_lists = [convert_to_number_list(line) for line in lines]
two_digit_numbers = [int(l[0]+l[-1]) for l in number_lists]
```

```python
sum(two_digit_numbers)
```
