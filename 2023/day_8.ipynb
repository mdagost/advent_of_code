## Part 1

```python
TEST_INFILE_1 = "inputs/day_8_test_1.txt"
TEST_INFILE_2 = "inputs/day_8_test_2.txt"
INFILE = "inputs/day_8_1.txt"

#with open(TEST_INFILE_1) as infile:
#with open(TEST_INFILE_2) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()

directions = lines[0]
nodes = lines[2:]
```

```python
directions, nodes
```

```python
nodes_dict = {}
for node in nodes:
    split = node.split("=")
    name = split[0].strip()
    lr = split[1].strip().replace("(", "").replace(")", "").split(",")
    left = lr[0].strip()
    right = lr[1].strip()
    nodes_dict[name] = {"L": left, "R": right}
```

```python
def get_direction():
    dirs = [d for d in directions]
    while True:
        dir = dirs[0]
        yield dirs[0]
        dirs = dirs[1:] + [dir]
        
```

```python
curr_node = "AAA"
steps = 0
for dir in get_direction():
    child_nodes = nodes_dict[curr_node]
    next_node = child_nodes[dir]
    steps += 1
    
    print(f"Step {steps}: {curr_node} -> {next_node} ({dir})")
    curr_node = next_node

    if curr_node == "ZZZ":
        break
```

## Part 2

```python
import math
from functools import reduce
```

```python
def get_freq(node):
    curr_node = node
    steps = 0
    for dir in get_direction():
        child_nodes = nodes_dict[curr_node]
        next_node = child_nodes[dir]
        steps += 1
        
        #print(f"Step {steps}: {curr_node} -> {next_node} ({dir})")
        curr_node = next_node

        if curr_node.endswith("Z"):
            return steps
```

```python
freqs = sorted([get_freq(n) for n in list(nodes_dict.keys()) if n.endswith("A")])
```

```python
freqs
```

```python
def least_common_multiple(a, b):
    return a*b // math.gcd(a, b)
```

```python
reduce(least_common_multiple, freqs)
```
