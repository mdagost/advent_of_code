## Part 1

```python
import math
```

```python
TEST_INFILE = "inputs/day_6_test_1.txt"
INFILE = "inputs/day_6_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()

times = [int(t) for t in lines[0].split(":")[1].strip().split()]
dists = [int(d) for d in lines[1].split(":")[1].strip().split()]
```

```python
"""
x * (T - x) > D
-x^2 + Tx - D > 0

roots are
-T + sqrt(T^2 - 4D) / -2 = (T - sqrt(T^2 - 4D))
-T - sqrt(T^2 - 4D) / -2 = (T + sqrt(T^2 - 4D))
"""
def get_wins(T, D):
    big_root   = (T + (T**2 - 4*D)**0.5) / 2.0
    small_root = (T - (T**2 - 4*D)**0.5) / 2.0

    big  = math.ceil(big_root - 1)
    small = math.floor(small_root + 1)

    wins = big - small + 1

    return int(wins)
```

```python
math.prod(get_wins(t, d) for t, d in zip(times, dists))
```
