## Part 1

```python
TEST_INFILE = "inputs/day_3_test_1.txt"
INFILE = "inputs/day_3_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
def parse_line(line_num, line):
    numbers = {}
    symbols = {}

    in_number = False
    for num, c in enumerate(line):
        # start of a number
        if not in_number and c.isnumeric():
            in_number = True
            number = c
            num_pos = [(line_num, num)]
            continue

        # continuing in a number
        if in_number and c.isnumeric():
            number += c
            num_pos.append((line_num, num))
            continue

        # end of number
        if in_number and not c.isnumeric():
            in_number = False
            numbers[tuple(num_pos)] = int(number)

            # still need to check for a symbol
            if c != ".":
                symbols[(line_num, num)] = c
            continue

        # check for a symbol
        if not in_number and not c.isnumeric():
            if c != ".":
                symbols[(line_num, num)] = c
            continue

    # if we're in a number at the end of a line end it
    if in_number:
        numbers[tuple(num_pos)] = int(number)

    return numbers, symbols
```

```python
def get_neighbors(pos, row_max, col_max):
    right_col = pos[1] + 1 if pos[1] + 1 < col_max else None
    left_col  = pos[1] - 1 if pos[1] - 1 >= 0 else None
    up_row    = pos[0] - 1 if pos[0] - 1 >= 0 else None
    down_row  = pos[0] + 1 if pos[0] + 1 < row_max else None

    neighbors = []
    neighbors.append((pos[0], right_col))
    neighbors.append((pos[0], left_col))
    neighbors.append((up_row, pos[1]))
    neighbors.append((down_row, pos[1]))

    neighbors.append((up_row, left_col))
    neighbors.append((up_row, right_col))
    neighbors.append((down_row, left_col))
    neighbors.append((down_row, right_col))

    neighbors = [n for n in neighbors if None not in n]

    return neighbors
```

```python
numbers, symbols = {}, {}
for num, line in enumerate(lines):
    num, sym = parse_line(num, line)
    numbers |= num
    symbols |= sym
```

```python
valid = {}

for symbol_pos in symbols.keys():
    symbol_neighbors = get_neighbors(symbol_pos, len(lines), len(lines[0]))

    for symbol_neighbor in symbol_neighbors:
        for number_positions in numbers.keys():
            if symbol_neighbor in number_positions:
                valid[number_positions] = numbers[number_positions]
```
