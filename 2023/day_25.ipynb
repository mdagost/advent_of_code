```python
import random
from collections import defaultdict
```

```python
TEST_INFILE = "inputs/day_25_test_1.txt"
INFILE = "inputs/day_25_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()
```

```python
graph = defaultdict(dict)
edges = []
for line in lines:
    e1 = line.split(":")[0]
    e2s = line.split(":")[1].strip().split(" ")
    print(f"{e1} -> {e2s}")
    for e2 in e2s:
        graph[e1][e2] = 1
        graph[e2][e1] = 1
        edges.append((e1, e2, f"{e1} <-> {e2}"))
```

```python
len(edges)
```

```python
def walk_graph(start_node, graph=graph):
    queue = [start_node]
    component = set()
    while len(queue) > 0:
        node = queue.pop(0)
        if node not in component:
            component.add(node)
            for neighbor in graph[node]:
                if neighbor not in component:
                    queue.append(neighbor)
    return component


def is_one_component(start_node, graph=graph):
    component = walk_graph(start_node, graph)
    return len(component) == len(graph)
```

```python
def get_random_edge(edges):
    return random.choice(edges)


def contract_edge(edge, edges):
    # we'll represent the contracted by by n_to_contract1
    n_to_contract1, n_to_contract2, n_to_contract_desc = edge
    new_edges = []
    for node1, node2, desc in edges:
        if node1 in [n_to_contract1, n_to_contract2] and node2 in [n_to_contract1, n_to_contract2]:
            continue
        elif node1 in [n_to_contract1, n_to_contract2]:
            new_edges.append((n_to_contract1, node2, desc))
        elif node2 in [n_to_contract1, n_to_contract2]:
            new_edges.append((node1, n_to_contract1, desc))
        else:
            new_edges.append((node1, node2, desc))
    
    return new_edges


def get_nodes(edges):
    nodes = set()
    for n1, n2, _ in edges:
        nodes.add(n1)
        nodes.add(n2)
    
    return nodes
```

```python
def karger(edges):
    working_edges = edges.copy()
    n_nodes = len(get_nodes(working_edges))
    while n_nodes > 2:
        working_edges = contract_edge(get_random_edge(working_edges), working_edges)
        n_nodes = len(get_nodes(working_edges))
    
    return working_edges
```

```python
working_edges = []
while len(working_edges) != 3:
    working_edges = karger(edges)
len(working_edges)
```

```python
for edge in working_edges:
    _, _, original_edge_desc = edge
    n1, n2 = original_edge_desc.split(" <-> ")
    del graph[n1][n2]
    del graph[n2][n1]
```
