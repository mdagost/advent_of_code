## Part 1

```python
TEST_INFILE = "inputs/day_13_test_1.txt"
INFILE = "inputs/day_13_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    puzzles = infile.read().split("\n\n")

puzzles_by_rows = [p.split() for p in puzzles]
puzzles_by_cols = [["".join(p[row][col] for row in range(len(p))) for col in range(len(p[0]))] for p in puzzles_by_rows]
```

```python
def folds_over_row(puzzle_by_rows, n):
    top = n
    bottom = top + 1

    if top == len(puzzle_by_rows) - 1:
        return False

    while top >= 0 and bottom < len(puzzle_by_rows):
        if puzzle_by_rows[top] != puzzle_by_rows[bottom]:
            return False
        top -= 1
        bottom += 1
    return True


def folds_over_col(puzzle_by_cols, n):
    left  = n
    right = left + 1

    if left == len(puzzle_by_cols) - 1:
        return False

    while left >= 0 and right < len(puzzle_by_cols):
        if puzzle_by_cols[left] != puzzle_by_cols[right]:
            return False
        left -= 1
        right += 1
    return True
```

```python
total = 0
for puzzle_n, puzzle in enumerate(puzzles_by_cols):
    for n in range(len(puzzle)):
        if folds_over_col(puzzle, n):
            print(f"Puzzle {puzzle_n} folds horizontally over col {n+1}")
            total += (n + 1) 

for puzzle_n, puzzle in enumerate(puzzles_by_rows):
    for n in range(len(puzzle)):
        if folds_over_row(puzzle, n):
            print(f"Puzzle {puzzle_n} folds vertically over row {n+1}")
            total += 100 * (n + 1)

total
```

## Part 2

```python
def smudge_folds_over_row(puzzle_by_rows, n):
    deltas = []

    top = n
    bottom = top + 1

    while top >= 0 and bottom < len(puzzle_by_rows):
        n_diff = sum(t != b for t, b in zip(puzzle_by_rows[top], puzzle_by_rows[bottom]))
        deltas.append(n_diff)
        top -= 1
        bottom += 1
    return deltas


def smudge_folds_over_col(puzzle_by_cols, n):
    deltas = []

    left  = n
    right = left + 1

    while left >= 0 and right < len(puzzle_by_cols):
        n_diff = sum(l != r for l, r in zip(puzzle_by_cols[left], puzzle_by_cols[right]))
        deltas.append(n_diff)
        left -= 1
        right += 1
    return deltas
```

```python
total = 0

for puzzle_n, puzzle in enumerate(puzzles_by_rows):
    for n in range(len(puzzle)):
        diffs = smudge_folds_over_row(puzzle, n)
        if len(diffs) and all(d in [0, 1] for d in diffs) and not all(d == 0 for d in diffs) and sum(diffs) == 1:
            print(f"Puzzle {puzzle_n} smudge folds vertically over row {n+1}")
            total += 100 * (n + 1) 

for puzzle_n, puzzle in enumerate(puzzles_by_cols):
    for n in range(len(puzzle)):
        diffs = smudge_folds_over_col(puzzle, n)
        if len(diffs) and all(d in [0, 1] for d in diffs) and not all(d == 0 for d in diffs) and sum(diffs) == 1:
            print(f"Puzzle {puzzle_n} smudge folds horizontally over col {n+1}")
            total += (n + 1)

total
```
