## Part 1

```python
from itertools import pairwise
```

```python
TEST_INFILE = "inputs/day_9_test_1.txt"
INFILE = "inputs/day_9_1.txt"

#with open(TEST_INFILE) as infile:
with open(INFILE) as infile:
    lines = infile.read().splitlines()

lines = [[int(e) for e in l.split()] for l in lines]
```

```python
lines
```

```python
def get_next_element(sequence):
    if all(s == 0 for s in sequence):
        return 0
    else:
        diffs = [e2 - e1 for e1, e2 in pairwise(sequence)]
        return sequence[-1] + get_next_element(diffs)
```

```python
sum(get_next_element(l) for l in lines)
```

## Part 2

```python
def get_previous_element(sequence):
    if all(s == 0 for s in sequence):
        return 0
    else:
        diffs = [e2 - e1 for e1, e2 in pairwise(sequence)]
        return sequence[0] - get_previous_element(diffs)
```

```python
sum(get_previous_element(l) for l in lines)
```
